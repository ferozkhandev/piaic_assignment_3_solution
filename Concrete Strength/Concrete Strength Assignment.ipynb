```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv("/content/compresive_strength_concrete.csv")
df.head()
```

```python
df.fillna(df.mean())
```

```python
df.columns = ["cement", "bfs", "fly_ash", "water", "superplasticizer", "coarse_aggregate", "fine_aggregate", "age", "ccs"]
df
```

```python
y = df['ccs']
df.drop('ccs', axis =1, inplace = True)
x = df.copy()
```

```python
y.head()
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
print(df.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train,y_train, test_size=0.2)
```

```python
print(df.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
print(x_val.shape)
print(y_val.shape)
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(x_train.shape)))
model.add(Dense(8, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))
```

```python
model.compile('Adam', loss='mse', metrics='mae')
```

```python
model.fit(x_train, y_train, epochs=100, verbose=1)
```

**Testing**

```python
print(x_test.shape, np.asarray(y_test.shape) )
```
